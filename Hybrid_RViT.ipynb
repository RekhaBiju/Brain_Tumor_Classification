```python
#import pillow

w=128
h=128
label_to_class = {
    'glioma': 0,
    'meningioma': 1,
    'notumor': 2,
    'pituitary':3


}
class_to_label = {v: k for k, v in label_to_class.items()}
n_classes = len(label_to_class)

def get_images(dir_name='data', label_to_class=label_to_class):
#read images / labels from directory

    Images = []
    Classes = []

    for j in ['/Training','/Testing']:
        for label_name in os.listdir(dir_name+str(j)):
            cls1 = label_to_class[label_name]

            for img_name in os.listdir('/'.join([dir_name+str(j), label_name])):
                # Load the image
                img = load_img('/'.join([dir_name+str(j), label_name, img_name]), target_size=(w, h))

                # Convert the image to a NumPy array
                img = img_to_array(img)

                Images.append(img)
                Classes.append(cls1)

    Images = np.array(Images, dtype=np.float16)
    Classes = np.array(Classes, dtype=np.float16)
    Images, Classes = shuffle(Images, Classes, random_state=0)

    return Images, Classes
```

```python
class ClassToken(tf.keras.layers.Layer):
    def __init__(self):
        super(ClassToken, self).__init__()

    def build(self, input_shape):
        w_init = tf.random_normal_initializer()
        self.w = self.add_weight(shape=(1, 1, input_shape[-1]), initializer=w_init, trainable=True )
       #self.w = tf.Variable(initial_value=w_init(shape=(1, 1, input_shape[-1]),dtype=np.float32), trainable=True )

    def call(self, inputs):
        batch_size = tf.shape(inputs)[0]
        hidden_dim = self.w.shape[-1]

        cls = tf.broadcast_to(self.w, [batch_size, 1, hidden_dim])
        cls = tf.cast(cls, dtype=inputs.dtype)

        return cls
```

```python
def mlp(x, cf):
    x = Dense(cf["mlp_dim"], activation="gelu")(x)
    x = Dropout(cf["dropout_rate"])(x)
    x = Dense(cf["hidden_dim"])(x)
    x = Dropout(cf["dropout_rate"])(x)

    return x
```

```python
def transformer_encoder(x, cf):
    skip_1 = x
    x = tf.keras.layers.LayerNormalization()(x)
    x = tf.keras.layers.MultiHeadAttention(num_heads=cf["num_heads"], key_dim=cf["hidden_dim"])(x, x)
    x = tf.keras.layers.Add()([x, skip_1])
    x = tf.keras.layers.Dropout(cf["dropout_rate"])(x)

    skip_2 = x
    x = tf.keras.layers.LayerNormalization()(x)
    x = mlp(x, cf)
    x = tf.keras.layers.Add()([x, skip_2])
    x = tf.keras.layers.Dropout(cf["dropout_rate"])(x)

    return x
```

```python
def ResNet50ViT(cf):
    input_shape = (cf["image_size"], cf["image_size"], cf["num_channels"])
    """ Input """
    inputs = layers.Input(shape=input_shape) ## (None, 128, 128, 3)

    """ Pre-trained Resnet50 """
    resnet50 = ResNet50(include_top=False, weights="imagenet", input_tensor=inputs)
    output = resnet50.output ## (None, 4, 4, 2048)

    """ Patch Embeddings """
    patch_embed = Conv2D(cf["hidden_dim"], kernel_size=cf["patch_size"], padding="same")(output) ##(None, 1, 1, 64)

    patch_embed = BatchNormalization()(patch_embed)
    _, h, w, f = patch_embed.shape
    patch_embed = Reshape((h*w, f))(patch_embed) ## (None, 1, 64)

    """ Position Embeddings """
    positions = tf.range(start=0, limit=cf["num_patches"], delta=1) ## (16,)
    pos_embed = Embedding(input_dim=cf["num_patches"], output_dim=cf["hidden_dim"])(positions) ## (16, 64)

    """ Patch + Position Embeddings """
    embed = patch_embed + pos_embed ## (None, 16, 64)

    """ Adding Class Token """
    token = ClassToken()(embed)
    x = Concatenate(axis=1)([token, embed]) ## (None, 17, 64)

    """ Transformer Encoder """
    for _ in range(cf["num_layers"]):
        x = transformer_encoder(x, cf)

    x = LayerNormalization()(x)
    x = x[:, 0, :]
    logits = Dense(cf["num_classes"])(x)

    model = Model(inputs, logits)
    model = keras.Model(inputs=inputs, outputs=logits)

    print(model.summary())
    return model
```

```python
# Required imports for prediction
from tensorflow.keras.applications import ResNet50
from tensorflow.keras import layers, Model
import keras

# Define class token if not already defined
class ClassToken(tf.keras.layers.Layer):
    def __init__(self):
        super(ClassToken, self).__init__()

    def build(self, input_shape):
        w_init = tf.random_normal_initializer()
        self.w = self.add_weight(shape=(1, 1, input_shape[-1]), initializer=w_init, trainable=True)

    def call(self, inputs):
        batch_size = tf.shape(inputs)[0]
        hidden_dim = self.w.shape[-1]
        cls = tf.broadcast_to(self.w, [batch_size, 1, hidden_dim])
        cls = tf.cast(cls, dtype=inputs.dtype)
        return cls

# Define helper functions if not already defined
def mlp(x, cf):
    x = layers.Dense(cf["mlp_dim"], activation="gelu")(x)
    x = layers.Dropout(cf["dropout_rate"])(x)
    x = layers.Dense(cf["hidden_dim"])(x)
    x = layers.Dropout(cf["dropout_rate"])(x)
    return x

def transformer_encoder(x, cf):
    skip_1 = x
    x = layers.LayerNormalization()(x)
    x = layers.MultiHeadAttention(num_heads=cf["num_heads"], key_dim=cf["hidden_dim"])(x, x)
    x = layers.Add()([x, skip_1])
    x = layers.Dropout(cf["dropout_rate"])(x)
    
    skip_2 = x
    x = layers.LayerNormalization()(x)
    x = mlp(x, cf)
    x = layers.Add()([x, skip_2])
    x = layers.Dropout(cf["dropout_rate"])(x)
    return x

def ResNet50ViT(cf):
    input_shape = (cf["image_size"], cf["image_size"], cf["num_channels"])
    inputs = layers.Input(shape=input_shape)
    
    resnet50 = ResNet50(include_top=False, weights="imagenet", input_tensor=inputs)
    output = resnet50.output
    
    patch_embed = layers.Conv2D(cf["hidden_dim"], kernel_size=cf["patch_size"], padding="same")(output)
    patch_embed = layers.BatchNormalization()(patch_embed)
    _, h, w, f = patch_embed.shape
    patch_embed = layers.Reshape((h*w, f))(patch_embed)
    
    positions = tf.range(start=0, limit=cf["num_patches"], delta=1)
    pos_embed = layers.Embedding(input_dim=cf["num_patches"], output_dim=cf["hidden_dim"])(positions)
    
    embed = patch_embed + pos_embed
    
    token = ClassToken()(embed)
    x = layers.Concatenate(axis=1)([token, embed])
    
    for _ in range(cf["num_layers"]):
        x = transformer_encoder(x, cf)
    
    x = layers.LayerNormalization()(x)
    x = x[:, 0, :]
    logits = layers.Dense(cf["num_classes"])(x)
    
    model = Model(inputs, logits)
    return model

# Define config if not already defined
if 'config' not in locals():
    config = {}
    config["num_layers"] = 6
    config["hidden_dim"] = 64
    config["mlp_dim"] = 2048
    config["num_heads"] = 8
    config["dropout_rate"] = 0.2
    config["image_size"] = 128
    config["patch_size"] = 32
    config["num_patches"] = int(config["image_size"]**2 / config["patch_size"]**2)
    config["num_channels"] = 3
    config["num_classes"] = 4
    w = 128
    h = 128

# Define class labels if not already defined
if 'class_to_label' not in locals():
    label_to_class = {
        'glioma': 0,
        'meningioma': 1,
        'notumor': 2,
        'pituitary': 3
    }
    class_to_label = {v: k for k, v in label_to_class.items()}

# Load the trained model
loaded_model = ResNet50ViT(config)
loaded_model.load_weights("hybrid.h5")

print("✓ Model loaded successfully!")
```

```python
def predict_on_image(image_path, model, class_to_label):
    """
    Predict the tumor type from a brain scan image.
    
    Args:
        image_path: Path to the image file
        model: Trained model
        class_to_label: Dictionary mapping class indices to class names
    
    Returns:
        predicted_class: Predicted tumor class name
        confidence: Confidence score (0-100%)
    """
    # Load and preprocess the image
    img = load_img(image_path, target_size=(w, h))
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_array = img_array.astype(np.float16)  # Match training data type
    
    # Make prediction
    logits = model.predict(img_array, verbose=0)
    predicted_class_idx = np.argmax(logits[0])
    confidence = float(np.max(logits[0]))
    
    # Get confidence as percentage (approximate)
    predicted_class = class_to_label[predicted_class_idx]
    
    return predicted_class, confidence, logits[0]
```

## Confusion Matrix for Predictions

```python
import glob
import seaborn as sns
import os
from sklearn.metrics import confusion_matrix, classification_report

# Get all test images from each category
all_test_images = []
true_labels = []

for class_name in ['glioma', 'meningioma', 'notumor', 'pituitary']:
    images = glob.glob(f"data/Testing/{class_name}/*")
    all_test_images.extend(images)
    true_labels.extend([class_name] * len(images))

print(f"Total test images: {len(all_test_images)}")

# Make predictions on all test images
y_true_names = []
y_pred_names = []

print("Making predictions on all test images...")
for i, image_path in enumerate(all_test_images):
    if (i + 1) % 50 == 0:
        print(f"  Processed {i + 1}/{len(all_test_images)}")
    
    predicted_class, confidence, _ = predict_on_image(image_path, loaded_model, class_to_label)
    
    # Extract true class from path - use os.path to handle both Windows and Unix paths
    true_class = os.path.basename(os.path.dirname(image_path))
    
    y_true_names.append(true_class)
    y_pred_names.append(predicted_class)

print("Predictions complete!")

# Encode labels to numbers
label_to_num = {'glioma': 0, 'meningioma': 1, 'notumor': 2, 'pituitary': 3}

# Convert to lowercase to handle case sensitivity
y_true_num = [label_to_num[label.lower()] for label in y_true_names]
y_pred_num = [label_to_num[label.lower()] for label in y_pred_names]

# Create confusion matrix
cm = confusion_matrix(y_true_num, y_pred_num, labels=[0, 1, 2, 3])

# Display confusion matrix
print("\nConfusion Matrix:")
print(cm)
```

```python
# Visualize confusion matrix as heatmap
plt.figure(figsize=(10, 8))

class_names = ['Glioma', 'Meningioma', 'No Tumor', 'Pituitary']
sns.heatmap(cm, 
            annot=True, 
            fmt='d', 
            cmap='Blues',
            xticklabels=class_names,
            yticklabels=class_names,
            cbar_kws={'label': 'Count'},
            square=True,
            linewidths=1,
            linecolor='black')

plt.title('Confusion Matrix - Brain Tumor Classification', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Predicted Label', fontsize=12, fontweight='bold')
plt.ylabel('True Label', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()

print("\nConfusion Matrix saved!")
```

```python
# Print detailed classification report
print("\n" + "="*70)
print("CLASSIFICATION REPORT")
print("="*70)
print(classification_report(y_true_num, y_pred_num, 
                          target_names=class_names,
                          digits=4))

# Calculate accuracy
accuracy = (cm.diagonal().sum()) / cm.sum()
print(f"Overall Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)")
print("="*70)
```
